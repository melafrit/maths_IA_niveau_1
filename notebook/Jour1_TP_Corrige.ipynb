```python
# 1. Vecteur du développeur #3
x3 = np.array([2, 2, 3, 200, 60])

# 2. Dimension
print(f"Dimension de x3 : {len(x3)}")  # ou x3.shape[0]

# 3. 3e composante (index 2)
niveau = x3[2]
print(f"Niveau d'études : {niveau}")

# 4. Norme
norme = np.sqrt(np.sum(x3 ** 2))
print(f"Norme de x3 : {norme:.2f}")

# 5. Développeur #5
x5 = np.array([0, 1, 1, 30, 50])

# 6. Produit scalaire
ps = x3 @ x5
print(f"Produit scalaire x3 · x5 = {ps}")

# 7. Distance
distance = np.sqrt(np.sum((x3 - x5) ** 2))
print(f"Distance : {distance:.2f}")
```

## Exercice 10 — Régression linéaire from scratch


```python
x = experience; y = salaire

x_mean = np.mean(x)
y_mean = np.mean(y)

numerateur = np.sum((x - x_mean) * (y - y_mean))
denominateur = np.sum((x - x_mean) ** 2)
w1 = numerateur / denominateur
w0 = y_mean - w1 * x_mean

y_pred = w1 * x + w0
mse = np.mean((y - y_pred) ** 2)

print(f"ŷ = {w1:.4f} × experience + {w0:.4f}")
print(f"MSE = {mse:.2f}, RMSE = {np.sqrt(mse):.2f} k€")

fig, ax = plt.subplots(figsize=(9, 5))
ax.scatter(x, y, c='blue', s=40, label='Données')
xl = np.linspace(0, 21, 100)
ax.plot(xl, w1*xl+w0, 'r-', lw=2, label=f'ŷ = {w1:.2f}x + {w0:.2f}')
ax.set_xlabel("Expérience"); ax.set_ylabel("Salaire (k€)")
ax.set_title(f"Régression (MSE={mse:.2f})"); ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

## Exercice 11 — Exploration complète


```python
# Statistiques descriptives
print(f"{'Colonne':<20} {'Moy':>8} {'Std':>8} {'Min':>8} {'Max':>8}")
print("-" * 55)
for j, nom in enumerate(noms_colonnes):
    col = data[:, j] if j < 6 else salaire_eleve
    print(f"{nom:<20} {col.mean():>8.1f} {col.std():>8.1f} {col.min():>8.1f} {col.max():>8.1f}")

# Corrélations
print("\n=== Corrélations avec le salaire ===")
correlations = []
for j in range(5):
    r = np.corrcoef(data[:, j], salaire)[0, 1]
    correlations.append(r)
    print(f"  {noms_colonnes[j]:<20} r = {r:+.3f}")

# 5 scatter plots
fig, axes = plt.subplots(1, 5, figsize=(18, 3.5))
for j in range(5):
    color = '#2ecc71' if abs(correlations[j]) > 0.5 else '#e74c3c'
    axes[j].scatter(data[:, j], salaire, c=color, s=30, alpha=0.7)
    axes[j].set_xlabel(noms_colonnes[j], fontsize=8)
    if j == 0: axes[j].set_ylabel("Salaire")
    axes[j].set_title(f"r = {correlations[j]:.2f}", fontsize=10)
    axes[j].grid(True, alpha=0.3)
plt.suptitle("Corrélations features vs salaire", fontweight='bold')
plt.tight_layout(); plt.show()

best = noms_colonnes[np.argmax(np.abs(correlations))]
worst = noms_colonnes[np.argmin(np.abs(correlations))]
print(f"\nPlus corrélée : {best}, Moins corrélée : {worst}")
```

## Exercice 12 — Comparer 3 modèles


```python
x = experience; y = salaire

# Modèle A : 1 feature (experience)
xm, ym = x.mean(), y.mean()
w1_A = np.sum((x-xm)*(y-ym)) / np.sum((x-xm)**2)
w0_A = ym - w1_A * xm
y_A = w1_A * x + w0_A
mse_A = np.mean((y - y_A)**2)

# Modèle B : 2 features, poids manuels
y_B = 2.0 * experience + 1.5 * nb_langages + 25
mse_B = np.mean((y - y_B)**2)

# Modèle C : 2 features, moindres carrés
X_aug = np.column_stack([experience, nb_langages, np.ones(n)])
w_C = np.linalg.lstsq(X_aug, y, rcond=None)[0]
y_C = X_aug @ w_C
mse_C = np.mean((y - y_C)**2)

print(f"{'Modèle':<12} {'Formule':<35} {'MSE':>8}")
print("-" * 58)
print(f"{'A (1 feat)':<12} {'ŷ = ' + f'{w1_A:.2f}×exp + {w0_A:.2f}':<35} {mse_A:>8.2f}")
print(f"{'B (manuel)':<12} {'ŷ = 2×exp + 1.5×lang + 25':<35} {mse_B:>8.2f}")
print(f"{'C (optimal)':<12} {'ŷ = ' + f'{w_C[0]:.2f}×exp + {w_C[1]:.2f}×lang + {w_C[2]:.2f}':<35} {mse_C:>8.2f}")
```
