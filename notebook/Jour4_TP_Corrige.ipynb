## Exercice 1 — Activations


```python
def sigmoid(z): return 1/(1+np.exp(-z))
def relu(z): return np.maximum(0, z)

valeurs = np.array([-3,-1,0,1,2,5])
print("sigmoid :", np.round(sigmoid(valeurs), 4))
print("relu    :", relu(valeurs))

z = np.linspace(-6, 6, 200)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(z, sigmoid(z), 'b-', lw=2); ax1.set_title("Sigmoïde"); ax1.grid(True, alpha=0.3)
ax2.plot(z, relu(z), 'r-', lw=2); ax2.set_title("ReLU"); ax2.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

## Exercice 2 — Forward pass


```python
def sigmoid(z): return 1/(1+np.exp(-z))
x = np.array([5.0, 4.0])
W1 = np.array([[0.3, 0.5], [-0.2, 0.4]]); b1 = np.array([-2.0, 0.5])
w2 = np.array([0.8, 0.6]); b2 = -0.5
z1 = W1 @ x + b1; h = sigmoid(z1); z2 = w2 @ h + b2; y_pred = sigmoid(z2)
print(f"h = {np.round(h, 4)}, ŷ = {y_pred:.4f}, classe {1 if y_pred>0.5 else 0}")
```

## Exercice 3 — XOR


```python
def sigmoid(z): return 1/(1+np.exp(-z))
def sigmoid_d(a): return a*(1-a)
X=np.array([[0,0],[0,1],[1,0],[1,1]],dtype=float); y=np.array([[0],[1],[1],[0]],dtype=float)
np.random.seed(42); W1=np.random.randn(2,4)*0.5; b1=np.zeros((1,4)); W2=np.random.randn(4,1)*0.5; b2=np.zeros((1,1))
for ep in range(5000):
    h=sigmoid(X@W1+b1); out=sigmoid(h@W2+b2); loss=np.mean((y-out)**2)
    do=(out-y)*sigmoid_d(out); W2-=1.0*(h.T@do)/4; b2-=1.0*np.mean(do,axis=0,keepdims=True)
    dh=(do@W2.T)*sigmoid_d(h); W1-=1.0*(X.T@dh)/4; b1-=1.0*np.mean(dh,axis=0,keepdims=True)
    if ep%1000==999: print(f"Epoch {ep+1}: loss={loss:.4f}, acc={np.mean((out>0.5).flatten()==y.flatten()):.0%}")
print(f"Prédictions : {out.flatten().round(3)}")
```

## Exercice 4 — Réseau sur salaire


```python
def sigmoid(z): return 1/(1+np.exp(-np.clip(z,-500,500)))
def sigmoid_d(a): return a*(1-a)

X_raw=data[:,:2]; y_cls=salaire_eleve.reshape(-1,1).astype(float)
mu=X_raw.mean(axis=0); std=X_raw.std(axis=0); X_n=(X_raw-mu)/std

np.random.seed(42); W1=np.random.randn(2,4)*0.5; b1=np.zeros((1,4)); W2=np.random.randn(4,1)*0.5; b2=np.zeros((1,1))
hist_loss=[]
for ep in range(2000):
    h=sigmoid(X_n@W1+b1); out=sigmoid(h@W2+b2); loss=np.mean((y_cls-out)**2); hist_loss.append(loss)
    do=(out-y_cls)*sigmoid_d(out); W2-=0.5*(h.T@do)/n; b2-=0.5*np.mean(do,axis=0,keepdims=True)
    dh=(do@W2.T)*sigmoid_d(h); W1-=0.5*(X_n.T@dh)/n; b1-=0.5*np.mean(dh,axis=0,keepdims=True)
    if ep%500==499: print(f"Epoch {ep+1}: loss={loss:.4f}, acc={np.mean((out>0.5)==y_cls):.0%}")

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(14,5))
ax1.plot(hist_loss,'b-'); ax1.set_title("Convergence"); ax1.grid(True,alpha=0.3)
xx,yy=np.meshgrid(np.linspace(X_n[:,0].min()-1,X_n[:,0].max()+1,100),np.linspace(X_n[:,1].min()-1,X_n[:,1].max()+1,100))
grid=np.c_[xx.ravel(),yy.ravel()]; zg=sigmoid(sigmoid(grid@W1+b1)@W2+b2).reshape(xx.shape)
ax2.contourf(xx,yy,zg,levels=20,cmap='RdYlGn',alpha=0.5); ax2.contour(xx,yy,zg,levels=[0.5],colors='black',linewidths=2)
for c,col in [(0,'#e74c3c'),(1,'#2ecc71')]:
    m=y_cls.flatten()==c; ax2.scatter(X_n[m,0],X_n[m,1],c=col,s=60,edgecolors='white')
ax2.set_title("Surface de décision"); plt.tight_layout(); plt.show()
```
