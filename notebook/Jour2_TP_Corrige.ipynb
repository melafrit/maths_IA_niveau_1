## Exercice 1 — GD 1D


```python
def J(w): return (w - 3) ** 2
def dJ(w): return 2 * (w - 3)

w = 0.0; alpha = 0.1
for i in range(20):
    grad = dJ(w)
    w = w - alpha * grad
    if i < 5 or i == 19:
        print(f"Iter {i+1:2d} : w = {w:.4f}, J = {J(w):.4f}")
print(f"\nRésultat : w = {w:.4f}")
```

## Exercice 2 — Visualiser


```python
def J(w): return (w-3)**2
def dJ(w): return 2*(w-3)
w=-2.0; alpha=0.15; hist_w=[w]; hist_J=[J(w)]
for i in range(25):
    w = w - alpha*dJ(w); hist_w.append(w); hist_J.append(J(w))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
wr = np.linspace(-3, 8, 200)
ax1.plot(wr, J(wr), 'b-', lw=2); ax1.plot(hist_w, hist_J, 'ro-', ms=5)
ax1.set_xlabel('w'); ax1.set_ylabel('J(w)'); ax1.grid(True, alpha=0.3)
ax2.plot(hist_J, 'g-o', ms=4); ax2.set_xlabel('Iter'); ax2.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

## Exercice 3 — Régression GD


```python
x = experience; y = salaire; n = len(y)
w1, w0 = 0.0, 0.0; alpha = 0.001; hist_mse = []
for it in range(5000):
    y_pred = w1*x + w0; errors = y - y_pred
    dw1 = -2/n * np.sum(x * errors)
    dw0 = -2/n * np.sum(errors)
    w1 -= alpha * dw1; w0 -= alpha * dw0
    hist_mse.append(np.mean(errors**2))
print(f"ŷ = {w1:.4f}x + {w0:.4f}, MSE = {hist_mse[-1]:.2f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
ax1.plot(hist_mse, 'b-'); ax1.set_title("Convergence"); ax1.grid(True, alpha=0.3)
ax2.scatter(x, y, c='blue', s=40); xl = np.linspace(0,21,100)
ax2.plot(xl, w1*xl+w0, 'r-', lw=2, label=f'{w1:.2f}x+{w0:.2f}')
ax2.legend(); ax2.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
```

## Exercice 4 — Comparer 4 alpha


```python
def J(w): return (w-5)**2
alphas = [0.001, 0.01, 0.1, 0.9]
colors = ['#e74c3c', '#f39c12', '#2ecc71', '#9b59b6']
fig, ax = plt.subplots(figsize=(8, 5))
for alpha, color in zip(alphas, colors):
    w = -3.0; hist = [J(w)]
    for _ in range(50):
        w = w - alpha * 2*(w-5); hist.append(J(w))
    ax.plot(hist, '-', color=color, lw=2, label=f'α={alpha}')
ax.set_xlabel("Itération"); ax.set_ylabel("J(w)")
ax.legend(); ax.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
```

## Exercice 5 — GD vs sklearn


```python
from sklearn.linear_model import LinearRegression

x = experience; y = salaire; n = len(y)
# GD
w1, w0 = 0.0, 0.0; alpha = 0.001
for _ in range(10000):
    err = y - (w1*x+w0); w1 -= alpha*(-2/n*np.sum(x*err)); w0 -= alpha*(-2/n*np.sum(err))
mse_gd = np.mean((y - w1*x - w0)**2)

# sklearn
model = LinearRegression(); model.fit(x.reshape(-1,1), y)
mse_sk = np.mean((y - model.predict(x.reshape(-1,1)))**2)

print(f"{'Méthode':<18} | {'w1':>8} | {'w0':>8} | {'MSE':>8}")
print("-" * 50)
print(f"{'Gradient (10k)':<18} | {w1:>8.4f} | {w0:>8.4f} | {mse_gd:>8.2f}")
print(f"{'sklearn':<18} | {model.coef_[0]:>8.4f} | {model.intercept_:>8.4f} | {mse_sk:>8.2f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
ax1.scatter(x, y, c='gray', s=30); xl = np.linspace(0,21,100)
ax1.plot(xl, w1*xl+w0, 'r-', lw=2, label=f'GD'); ax1.plot(xl, model.predict(xl.reshape(-1,1)), 'g--', lw=2, label='sklearn')
ax1.legend(); ax1.grid(True, alpha=0.3); ax1.set_title("Droites de régression")
plt.tight_layout(); plt.show()
```
