## 1. L'IA comme fonction mathématique

> **Intuition** : Un modèle d'IA est une fonction $f$ qui prend des données en entrée et produit une prédiction en sortie.

$$f_{\theta}(\mathbf{x}) = \hat{y}$$

- $\mathbf{x}$ : les **features** (caractéristiques d'un développeur)
- $\hat{y}$ : la **prédiction** (salaire estimé)
- $\theta$ : les **paramètres** que le modèle apprend


---
## 2. Vecteurs : représenter un développeur

> **Définition** : Un **vecteur** $\mathbf{x} \in \mathbb{R}^p$ est une liste ordonnée de $p$ nombres.

Chaque développeur est un vecteur de 5 features :

$$\mathbf{x} = \begin{pmatrix} \text{experience} \\ \text{nb\_langages} \\ \text{niveau\_etudes} \\ \text{taille\_entreprise} \\ \text{remote} \end{pmatrix}$$


```python
# Vecteur du développeur #2 (index 2)
x2 = data[2, :5]  # on exclut le salaire (c'est la cible)
print(f"Développeur #2 : {x2}")
print(f"  Experience     : {x2[0]} ans")
print(f"  Langages       : {x2[1]}")
print(f"  Niveau études  : {x2[2]} (4 = Bac+5)")
print(f"  Taille entreprise : {x2[3]} employés")
print(f"  Remote         : {x2[4]}%")
print(f"\nDimension du vecteur : {x2.shape[0]} (on dit x ∈ ℝ⁵)")
```

---
## 3. Matrices : le dataset complet

> **Définition** : Une **matrice** $\mathbf{X} \in \mathbb{R}^{n \times p}$ empile $n$ vecteurs de dimension $p$.

Notre dataset est une matrice $30 \times 5$ (30 développeurs, 5 features).


```python
# Matrice des features (sans le salaire)
X = data[:, :5]
print(f"Matrice X : shape = {X.shape}")
print(f"  n = {X.shape[0]} observations (développeurs)")
print(f"  p = {X.shape[1]} features")
print(f"\n3 premières lignes :")
print(X[:3])

# Slicing : extraire une colonne entière
print(f"\nToutes les expériences : {X[:, 0]}")
```

---
## 4. Produit scalaire

> **Définition** : Le produit scalaire de $\mathbf{x}$ et $\mathbf{w}$ (même dimension $p$) :
>
> $$\langle \mathbf{x}, \mathbf{w} \rangle = \mathbf{x}^T \mathbf{w} = \sum_{j=1}^{p} x_j \cdot w_j$$

**Intuition** : c'est une **moyenne pondérée**. Les poids $\mathbf{w}$ disent l'importance de chaque feature.


```python
# Produit scalaire = prédiction du modèle
x = np.array([5, 4])          # un dev : 5 ans exp, 4 langages
w = np.array([2.0, 1.5])      # poids : importance de chaque feature

# 3 façons de calculer le produit scalaire
ps1 = np.sum(x * w)           # méthode 1 : sum(element-wise)
ps2 = np.dot(x, w)            # méthode 2 : np.dot
ps3 = x @ w                   # méthode 3 : opérateur @ (recommandé)

print(f"x = {x}, w = {w}")
print(f"Produit scalaire : {ps1} = {ps2} = {ps3}")
print(f"Prédiction : ŷ = x @ w + b = {ps3} + 25 = {ps3 + 25} k€")
```

---
## 5. Multiplication matrice-vecteur

> $$\hat{\mathbf{y}} = \mathbf{X} \mathbf{w}$$
>
> Calculer $n$ produits scalaires **en parallèle** : chaque ligne de $\mathbf{X}$ est multipliée par $\mathbf{w}$.


```python
# Prédiction pour TOUS les développeurs d'un coup
X_2feat = data[:, :2]  # experience + nb_langages
w = np.array([2.0, 1.5])
b = 25.0

y_pred = X_2feat @ w + b  # 30 prédictions en une ligne !
y_vrai = salaire

print("Dev | Exp | Lang | Prédit | Vrai | Erreur")
print("-" * 48)
for i in range(8):
    err = y_vrai[i] - y_pred[i]
    print(f" {i:2d} | {X_2feat[i,0]:3.0f} | {X_2feat[i,1]:4.0f} | {y_pred[i]:6.1f} | {y_vrai[i]:4.1f} | {err:+5.1f}")
```

---
## 6. Régression linéaire

> **Modèle** : $\hat{y} = w_1 \cdot x_{\text{exp}} + w_0$
>
> **MSE** (Mean Squared Error) : $J = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$


```python
# Régression linéaire : expérience → salaire
x = experience
y = salaire

# Formule analytique des moindres carrés
x_mean, y_mean = x.mean(), y.mean()
w1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
w0 = y_mean - w1 * x_mean

y_pred = w1 * x + w0
mse = np.mean((y - y_pred)**2)

print(f"Droite de régression : ŷ = {w1:.4f} × experience + {w0:.4f}")
print(f"MSE = {mse:.2f}")
print(f"RMSE = {np.sqrt(mse):.2f} k€ (erreur typique)")

# Graphique
fig, ax = plt.subplots(figsize=(9, 5))
ax.scatter(x, y, c='#3b82f6', s=50, label='Données')
xl = np.linspace(0, 21, 100)
ax.plot(xl, w1*xl+w0, 'r-', lw=2, label=f'ŷ = {w1:.2f}x + {w0:.2f}')
ax.set_xlabel("Expérience (années)"); ax.set_ylabel("Salaire (k€)")
ax.set_title("Régression linéaire — Jour 1"); ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

## ✅ Résumé du Jour 1

| Concept | Formule | Python |
|---------|---------|--------|
| Vecteur | $\mathbf{x} \in \mathbb{R}^p$ | `np.array([...])` |
| Matrice | $\mathbf{X} \in \mathbb{R}^{n \times p}$ | `data[:, :5]` |
| Produit scalaire | $\mathbf{x}^T \mathbf{w}$ | `x @ w` |
| Prédiction | $\hat{y} = w_1 x + w_0$ | `w1 * x + w0` |
| MSE | $\frac{1}{n}\sum(y_i - \hat{y}_i)^2$ | `np.mean((y - y_pred)**2)` |
